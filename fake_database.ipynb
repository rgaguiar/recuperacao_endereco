```python
import pandas as pd
import random
from faker import Faker
```

```python
# Inicializar Faker com suporte ao português
fake = Faker("pt_BR")
```

```python
# Função para gerar dados da tabela 'Clientes'
def criar_clientes(n=1000):
    genero = ["Masculino", "Feminino"]
    estado_civil = ["Solteiro", "Casado", "Divorciado", "Viúvo"]
    empregado = ["Sim", "Não"]
    nivel_educacao = ["Ensino Fundamental", "Ensino Médio", "Ensino Superior", "Pós-graduação"]

    clientes = []
    for i in range(1, n+1):
        clientes.append({
            "cliente_id": i,
            "nome": fake.name(),
            "data_nascimento": fake.date_of_birth(minimum_age=20),
            "email": fake.email(),
            "telefone": fake.phone_number(),
            "data_cadastro": fake.date_between(start_date="-1y", end_date="today"),
            "genero": random.choice(genero),
            "estado_civil": random.choice(estado_civil),
            "profissao": fake.job(),
            "empregado": random.choice(empregado),
            "renda_mensal": round(random.uniform(1000, 100000), 2),
            "numero_filhos": random.randint(0, 5),
            "nivel_educacao": random.choice(nivel_educacao),
            "endereco": fake.address(),
            "cidade": fake.city(),
            "estado": fake.state_abbr(),
            "cep": fake.postcode()
        })
    return pd.DataFrame(clientes)

clientes_df = criar_clientes()

clientes_df['data_nascimento'] = pd.to_datetime(clientes_df['data_nascimento'])
clientes_df['data_cadastro'] = pd.to_datetime(clientes_df['data_cadastro'])
clientes_df.info()
display(clientes_df)
```

```python
# Função para criar dados da tabela produtos
def criar_produtos(n=100):
    categorias = ["Casual", "Formal", "Semi-formal", "Criativo", "Esportivo"]
    tipos_de_roupas = [
    "Calça jeans",
    "Camiseta gola V",
    "Camisa social",
    "Jaqueta",
    "Vestido",
    "Blusa de frio",
    "Shorts",
    "Saia",
    "Macacão",
    "Blazer",
    "Camiseta regata",
    "Moletom",
    "Terno",
    "Polo",
    "Bermuda"
]
    produtos = []
    for i in range(1, n+1):
        nome_produto = random.choice(tipos_de_roupas)
        produtos.append({
            "produto_id": i,
            "nome": random.choice(tipos_de_roupas),
            "categoria": random.choice(categorias),
            "preco_unitario": round(random.uniform(10, 2000), 2),
            "estoque_atual": random.randint(0, 500)
        })
    return pd.DataFrame(produtos)

produtos_df = criar_produtos(5)
produtos_df
```

```python
# Função para gerar dados da tabela 'Lojas'
def criar_lojas(n=10):
    lojas = []
    for i in range(1, n+1):
        lojas.append({
            "loja_id": i,
            "nome": f"Loja {i}",
            "endereco": fake.address(),
            "cidade": fake.city(),
            "estado": fake.state_abbr()
        })
    return pd.DataFrame(lojas)
lojas_df = criar_lojas(5)
lojas_df
```

```python
# Função para gerar dados da tabela 'Colaboradores'
def criar_colaboradores(n=50):
    cargos = ["Estagiário", "Vendedor Junior", "Vendedor Pleno", "Vendedor Senior", "Gerente"]
    colaboradores = []
    for i in range(1, n+1):
        loja = random.choice(lojas_df["loja_id"])
        colaboradores.append({
            "colaborador_id": i,
            "loja_id": loja,
            "nome": fake.name(),
            "cargo": random.choice(cargos),
            "data_admissao": fake.date_between(start_date="-1y", end_date="today")

        })

    # Definir os salários padrão para cada cargo

    return pd.DataFrame(colaboradores)

colaboradores_df = criar_colaboradores(50)

salarios_por_cargo = {
    "Estagiário": 1000.00,
    "Vendedor Junior": 1500.00,
    "Vendedor Pleno": 2500.00,
    "Vendedor Senior": 4000.00,
    "Gerente": 6000.00
}

# Atualizar os salários com base no cargo

colaboradores_df['salario'] = colaboradores_df['cargo'].map(salarios_por_cargo)
colaboradores_df['data_admissao'] = pd.to_datetime(colaboradores_df['data_admissao'])
display(colaboradores_df.head())
colaboradores_df.info()
```

```python
# Função para gerar dados da tabela 'Vendas'
def criar_vendas(clientes_df, colaboradores_df, lojas_df, produtos_df, n=100000):
    vendas = []
    for i in range(1, n+1):
        cliente = random.choice(clientes_df["cliente_id"])
        vendedor = random.choice(colaboradores_df["colaborador_id"])
        loja = random.choice(lojas_df["loja_id"])
        produto = random.choice(produtos_df["produto_id"])
        quantidade = random.randint(1, 10)
        preco_unitario = produtos_df.loc[produtos_df["produto_id"] == produto, "preco_unitario"].values[0]
        vendas.append({
            "venda_id": i,
            "cliente_id": cliente,
            "vendedor_id": vendedor,
            "loja_id": loja,
            "produto_id": produto,
            "quantidade": quantidade,
            "preco_total": round(preco_unitario * quantidade, 2),
            "data_venda": fake.date_between(start_date="-1y", end_date="today")
        })
    return pd.DataFrame(vendas)

vendas_df = criar_vendas(clientes_df, colaboradores_df, lojas_df, produtos_df, n=100000)

vendas_df["data_venda"] = pd.to_datetime(vendas_df["data_venda"])
vendas_df.info()
```
